```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2,weights="distance")
```

```python
df["Married"]=df["Married"].round(decimals=0)
```

```python
df["Self_Employed"]=df["Self_Employed"].round(decimals=0)
```

```python
df["Dependents"]=imputer.fit_transform(df["Dependents"].values.reshape(-1,1))
```

```python
df["Dependents"]=df["Dependents"].round(decimals=0)
```

```python
df["Credit_History"].value_counts()
```

```python
imputer = KNNImputer(n_neighbors=2,weights="distance")
```

```python
df["Credit_History"]=imputer.fit_transform(df["Credit_History"].values.reshape(-1,1))
```

```python
df.isnull().sum()
```

```python
df["LoanAmount"].fillna(df["LoanAmount"].mean(),inplace=True)
```

```python
df["Loan_Amount_Term"].fillna(df["Loan_Amount_Term"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
X_train=df.drop(columns=["Loan_ID","Loan_Status"],axis=1)
```

```python
y_train=df["Loan_Status"]
```

```python
model.fit(X_train,y_train)
```

```python
y_trainpreds=model.predict(X_train)
```

```python
print(y_trainpreds)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_trainpreds,y_train)
```

```python
print(accuracy)
```
